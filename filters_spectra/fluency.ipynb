```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
spectra1 = '80kV42deg1Air0.5Be0.5Ag0.5Mo.txt'
spectra2 = '82kV42deg1Air0.5Be0.5Ag0.5Mo.txt'

datos1 = np.loadtxt(spectra1, delimiter='\t')
x1 = datos1[:, 0]  
y1 = datos1[:, 1]  


datos2 = np.loadtxt(spectra2, delimiter='\t')
x2 = datos2[:, 0]  
y2 = datos2[:, 1]  


plt.figure(figsize=(10, 6))


plt.plot(x1, y1, marker='o', linestyle='-', color='b', label='80Kv') 


plt.plot(x2, y2, marker='x', linestyle='--', color='r', label='82Kv')  


plt.title('Comparación de fluencia de fotones')
plt.xlabel('Energy (keV)')
plt.ylabel('Fluence (Photons/mm^2)')
plt.legend()  

plt.ylim(0, 3.7e6)

plt.grid(True)  
plt.show()
```
